# Research Question

How has Netflix's content library evolved over the years?

This analysis aims to answer the following questions:

- Which type of content is more common on Netflix (Movies or TV Shows)?
- Which countries contribute the most Netflix content?
- How has Netflix's content production changed over the years?
- What are the most common content ratings on Netflix?

```python
import matplotlib.pyplot as plt

content_type = df['type'].value_counts()

plt.figure(figsize=(6,4))
content_type.plot(kind='bar', color=['skyblue', 'orange'])

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(12,5))

df['release_year'].value_counts().sort_index().plot()

plt.title("Netflix Content Released Over the Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")

plt.show()
```

```python
top_countries = df['country'].dropna().value_counts().head(10)

plt.figure(figsize=(10,5))
top_countries.plot(kind='bar')

plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")

plt.show()
```

```python
plt.figure(figsize=(10,5))

df['rating'].value_counts().plot(kind='bar')

plt.title("Distribution of Netflix Content Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")

plt.show()
```

# Conclusion

### Key Findings

- Movies are much more common than TV Shows on Netflix.
- The United States contributes the highest number of Netflix titles, followed by India.
- Netflix content increased rapidly after 2015, reaching its peak around 2019–2020.
- TV-MA and TV-14 are the most common content ratings, indicating that Netflix mainly focuses on mature audiences.

### Overall Insight

The analysis shows that Netflix has expanded its content library significantly over the years, with a strong focus on movies, international content, and adult-oriented entertainment.
